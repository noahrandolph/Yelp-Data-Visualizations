```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

business=pd.read_csv("yelp_business.csv")

%matplotlib inline
```

#### Which companies have the most entries in Yelp?

```python
print(business['name'].value_counts()[:50])
```

#### What cities are represented?

```python
print(business['city'].value_counts())
```

#### Neighborhood a useful column or too many nulls?

```python
pd.isnull(business.neighborhood).value_counts()
```

#### Which locations have the most reviews?

```python
business.sort_values(by=['review_count'], ascending=False)
```

#### What about review counts of businesses *not* in Las Vegas?

```python
business[business.city != "Las Vegas"].sort_values(by=['review_count'], ascending=False)
```

#### What's in the column "categories?"

```python
print(business['categories'].value_counts()[:50])
```

```python
print(business['categories'].value_counts()[-50:])
```

#### What's in the is_open column?

```python
print(business['is_open'].value_counts()[:50])
```

#### Just breweries

```python
breweries = business['categories'].str.extract('(?P<brew>(Brew))', expand=True)
brew_biz = business[breweries.notnull()['brew']]

brew_biz.sort_values(by=['review_count'], ascending=False)
```

#### Just Vegas Starbucks

```python
x = business[business.city == "Las Vegas"]['stars'].value_counts()
x = x.sort_index()
x.plot(kind='bar')
#plot
plt.title("Las Vegas Star Rating Distribution")
plt.ylabel('# of businesses', fontsize=12)
plt.xlabel('Star Ratings ', fontsize=12)
```
